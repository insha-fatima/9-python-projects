```python
import time
def count_down(seconds):
  while seconds > 0:
    minutes, seconds_remaining = divmod(seconds, 60)
    seconds_display = seconds_remaining if minutes == 0 else f"{minutes}:{seconds_remaining:02d}"
    print(f"00:{seconds_remaining:02d}" if minutes == 0 else f"{minutes}:{seconds_remaining:02d}", end='\r')
    time.sleep(1)
    seconds -= 1
  print("00:00")
total_time = int(input("Enter time in second for countdown: "))
count_down(total_time)
print("Your time up!")

```
