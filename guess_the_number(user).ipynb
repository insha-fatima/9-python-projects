```python
import random
print("Guess your number between 1 to 100")
number = random.randint(1,100)
guess = int(input("Enter your guess number!"))

while True:
  guess = int(input("Enter your guess number!"))
  if guess < number:
    print("To small number")
  elif guess > number:
    print("To large number")
  else:
    print("Congratulations! Your guess number is correct")
  break
```
