```python
import random

words = ["python", "vscode", "game", "string", "boolean", "tupple"]
word = random.choice(words)
guessed_letters = []
tries = 6

print("Welcome to Hangman game")
print("_ " * len(word))

while tries > 0:
    guess = input("\n Guess the letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Write one alphabet only!")
        continue

    if guess in guessed_letters:
        print("This letter you have already chosen, let's choose another letter.")
        continue

    guessed_letters.append(guess)

    if guess in word:
        print("You chose a correct letter!")
    else:
        tries -= 1
        print(f"Wrong! {tries} tries left.")

    displayed_word = " ".join([letter if letter in guessed_letters else "_" for letter in word])
    print(displayed_word)

    if "_" not in displayed_word:
        print(f"Congratulations! The correct word is: {word}")
        break
else:
    print(f"Game over! The correct word is: {word}")

```
